```python
import pandas as pd
import numpy as np
```

```python
dict1={"Roll_No":[1,2,3,4,5],"Name":["Amol","Dipak","Shreya","Krisha","Pooja"],"Maths_marks":[60,70,80,90,np.nan],"English_marks":[70,80,90,np.nan,60],"Science_marks":[800,90,np.nan,60,70],"History_marks":[90,np.nan,60,70,80],"Geography_marks":[np.nan,60,70,80,90]}
```

```python
df=pd.DataFrame(dict1)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
#1
# Handling missing values
# For numerical columns, replace missing values with the column mean
numeric_cols = df.select_dtypes(include=np.number).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())
```

```python
df
```

```python
# Handling inconsistencies
# Convert categorical values to lowercase
categorical_cols = df.select_dtypes(include='object').columns
for col in categorical_cols:
    df[col] = df[col].str.lower()

# Display the DataFrame after handling inconsistencies
print("\nDataFrame after handling inconsistencies:")
print(df)
```

```python
#2
# Detecting and handling outliers
# Define a function to detect outliers using the Interquartile Range (IQR) method
def detect_outliers_iqr(data):
    Q1 = np.nanpercentile(data, 25)
    Q3 = np.nanpercentile(data, 75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return (data < lower_bound) | (data > upper_bound)
```

```python
# Scan numeric variables for outliers using the IQR method
numeric_cols = df.select_dtypes(include=np.number).columns
outliers_mask = df[numeric_cols].apply(detect_outliers_iqr)

outliers_mask
```

```python
# Deal with outliers by replacing them with the median value
for col in numeric_cols:
    median = df[col].median()
    df.loc[outliers_mask[col], col] = median
```

```python
df
```

```python
#3
# Apply data transformation (log transformation) to decrease skewness
# We'll apply log transformation to the "Maths_marks" variable
df['Transformed_Maths_marks'] = np.log(df['Maths_marks'])
```
